```python
# Obtener el proyecto actual
project_id = os.environ.get("Proyecto Final - Henry")

# Crear un cliente de Cloud Storage
client = storage.Client(project=project_id)

# Obtener el nombre del bucket
bucket_name = "gmy"

# Obtener el objeto de referencia al bucket
bucket = client.bucket(bucket_name)

# Crear un DataFrame vacío
df = pd.DataFrame()

# Iterar sobre los blobs
for numero_archivo in range(2, 12):
    # Obtener la lista de blobs
    blob = bucket.get_blob(f"metadata-sitios/{numero_archivo}.json")
    
    # Descargar el blob
    data = blob.download_as_string()

    # Convertir el objeto de bytes a un objeto similar a un archivo
    file = io.BytesIO(data)

    # Leer el archivo JSON en un DataFrame
    df_temp = pd.read_json(file, lines=True)

    # Seleccionar las columnas deseadas
    df_temp = df_temp[['gmap_id', 'name', 'address', 'latitude', 'longitude']]

    # Agregar las nuevas filas al DataFrame
    df = pd.concat([df, df_temp], ignore_index=True)

# Guardar el DataFrame en un CSV
df.to_csv(bucket.blob("etl/metadata-sitios.csv").open("w"), index=False)
```

```python
# Obtener el proyecto actual
project_id = os.environ.get("Proyecto Final - Henry")

# Crear un cliente de Cloud Storage
client = storage.Client(project=project_id)

# Obtener el nombre del bucket
bucket_name = "gmy"

# Obtener el objeto de referencia al bucket
bucket = client.bucket(bucket_name)

# Obtener una lista de las carpetas de los estados
folders = [x for x in bucket.list_blobs(prefix="reviews-estados/")]

# Crear un DataFrame vacío
df = pd.DataFrame()

# Iterar sobre las carpetas de los estados
for folder in folders:

    # Obtener el nombre del estado
    state = folder.name.split("/")[1]
    state = state.split("-")
    state = state[1]
    filename = folder.name.split("/")[-1]
    print(folder.name, state, filename)

    # Obtener el blob
    blob = bucket.get_blob(folder.name)

    # Descargar el blob
    data = blob.download_as_string()

    # Convertir el objeto de bytes a un objeto similar a un archivo
    file = io.BytesIO(data)

    # Leer el archivo JSON en un DataFrame
    df_review = pd.read_json(file, lines=True)
    
    # Eliminar las columnas pics y resp
    df_review = df_review.drop(["name","text","pics", "resp"], axis=1)

    # Eliminar los que tengan la columna text nula
    df_review = df_review[df_review["rating"].notnull()]
    
    # Agregar la columna Estado
    df_review["Estado"] = state

    # Agregar los datos al DataFrame
    df = pd.concat([df, df_review], ignore_index=True)    
    
# Guardar el DataFrame
df.to_csv(bucket.blob("etl/reviews-estados.csv").open("w"), index=False)
```
