## Recorrer la carpeta metadata-sitiosy extraer campos de los 11 archivos JSon

```python
# Crear un DataFrame vacío
df_gmms = pd.DataFrame()

# Iterar sobre los blobs
for numero_archivo in range(1, 12):
    # Obtener la lista de blobs
    blob = bucket.get_blob(f"metadata-sitios/{numero_archivo}.json")
    print(f"metadata-sitios/{numero_archivo}.json")
    
    # Descargar el blob
    data = blob.download_as_string()

    # Convertir el objeto de bytes a un objeto similar a un archivo
    file = io.BytesIO(data)

    # Leer el archivo JSON en un DataFrame
    df_temp = pd.read_json(file, lines=True)

    # Seleccionar las columnas deseadas
    df_temp = df_temp[['gmap_id', 'name', 'address', 'latitude', 'longitude','avg_rating','num_of_reviews','state']]

    # Agregar las nuevas filas al DataFrame
    df_gmms = pd.concat([df_gmms, df_temp], ignore_index=True)

# Guardar el DataFrame en un CSV
df_gmms.to_csv(bucket.blob("etl/metadata-sitios.csv").open("w"), index=False)
```

```python
# Guardar el DataFrame en un parquet
df_gmms.to_parquet(bucket.blob("etl/metadata-sitios.parquet").open("wb"))
```

## Extraer campo MISC de los 11 archivos JSon

```python
# Crear un DataFrame vacío
df_MISC = pd.DataFrame()
print('Inicio MISC')
# Iterar sobre los blobs
for numero_archivo in range(1, 12):
    # Obtener la lista de blobs
    blob = bucket.get_blob(f"metadata-sitios/{numero_archivo}.json")
    print(f"metadata-sitios/{numero_archivo}.json")
    
    # Descargar el blob
    data = blob.download_as_string()

    # Convertir el objeto de bytes a un objeto similar a un archivo
    file = io.BytesIO(data)

    # Leer el archivo JSON en un DataFrame
    df_temp = pd.read_json(file, lines=True)

    # Seleccionar las columnas deseadas
    df_temp = df_temp[['gmap_id','MISC']]

    # Agregar las nuevas filas al DataFrame
    df_MISC = pd.concat([df_MISC, df_temp], ignore_index=True)

# Guardar el DataFrame en un CSV
df_MISC.to_csv(bucket.blob("etl/metadata-sitios-MISC.csv").open("w"), index=False)
# Guardar el DataFrame en un parquet
df_MISC.to_parquet(bucket.blob("etl/metadata-sitios-MISC.parquet").open("wb"))
print('Fin MISC')
```

```python
# Crear un DataFrame vacío
df_desanidar = pd.DataFrame()

# Convertir los diccionarios en cadenas JSON
df_desanidar["MISC"] = df_MISC["MISC"].apply(json.dumps)

# Cargar las cadenas JSON con la función json.loads()
df_desanidar["MISC"] = df_desanidar["MISC"].apply(json.loads)

df_desanidar
```

```python
# Desanidar el DataFrame
def desanidar_dataframe(df):
    """Desanida un DataFrame.

    Args:
        df: Un DataFrame.

    Returns:
        Un DataFrame desanidado.
    """

    result = pd.DataFrame()
    for i in range(len(df)):
        row = df.iloc[i]
        for key, value in row["MISC"].items():
            result.loc[i, key] = value
    return result
```

```python
df_desanidar["MISC"] = desanidar_dataframe(df_desanidar["MISC"])

# Imprimir el DataFrame desanidado
print(df_desanidar["MISC"])
```
